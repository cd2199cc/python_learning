## 循环

### while 循环

while condition:

    statement(s)

```python
# 循环语句break：跳出循环，执行循环以后的语句

while True:
    answer = input('can i stop while ??')
    if answer == 'ok':
        break
    print('i am tired!!')
```

```python
# 循环语句continue：跳出本次循环，回到循环开头

i=0
while i<10:
    i+=1
    if i%2==0:
        continue
    print(i)
        
```

```python
# while else 只有在执行完while才会执行else后面的，若中途break则不执行else以后的

i=0
while i<3:
    i+=1
    answer=input('今天打卡了吗？')
    if answer == 'yes':
        continue
    else:
        break
else:
    print('congraduation!')
            
```

```python
#打印九九乘法表
i=1
while i<10:
    j=1
    while j<=i:
        print(j, '*', i, '=', i*j, end=' ')   # end='？' 表示不换行，在这一行的后面加一个？后继续打印
        j+=1
    print()
    i+=1
```

### 列表的方法
添加元素：append‘只能添加一个’, extend‘可以添加多个’, insert‘中间插入’
删除元素：remove‘删除指定元素’，pop‘按索引删除’，clear‘清空列表’
sort()‘给数字列表从小到大排序’reverse（）‘反转列表’

```python
aa = [1, 2, 3, 4, 'haha', '@2_','**',] 
aa.extend([33, 'abs'])  # 添加33和abs两个元素
aa[len(aa):] = ['hahaha', 45]  #添加
aa.insert(2,'iii')   # 在第二个元素后面插入iii
aa.remove('**')  # 删除第一个**
bb = aa.copy()   # 把aa列表复制一份并命名为bb
aa.count[2]   # 数一下aa中有几个2

# is语句（判断两个变量是否指向同一个对象），a is b 如果a, b指向同一个对象，则返回true，否则返回false
aa is bb   # 判断aa和bb是不是同一个对象
```

### 列表的嵌套（多维列表）

```python
# 创建一个二维列表
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]  # 相当于一个二维矩阵

# 遍历列表的元素
for i in matrix:
    for j in i:
        print(j, end = ' ')
    print()
    
# 列表元素的查找
matrix[0][0]
```

```python
# 多维列表的复制需要用到copy库的deepcopy函数
import copy
y = copy.copy(matrix)
y[1][1] is matrix[1][1]
```

### 列表推导式

基本规范  [expression for target in iterable]

简化了for循环的使用

```python
# 利用for循环来进行
i = [1, 2, 3, 4]
for j in range(len(i)):
    i[j] +=1
i
```

```python
# 利用列表推导式
a = [1, 2, 3, 4]
a = [j+1 for j in a]
a
```

```python
# 利用列表推导式提取矩阵的某一列元素
matric = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
a = [matric[i][1] for i in range(3)]
a

# 获取matric主对角线的元素
b = [matric[i][i] for i in range(len(matric))]
b

# 获取matric副对角线的元素
c = [matric[i][len(matric)-1-i] for i in range(len(matric))]
c
```

>高阶应用：
1.创建高阶列表
2.加入if语句
3.循环嵌套


>[expression for i_1 in m_1 if condition_1

              for i_2 in m_2 if condition_2
        
                    ······
              
              for i_n in m_n if condotion_n]

```python
# 创建二维列表
a = [[0] * 3 for i in range(3)]
a
```

```python
# 加入if条件句
b = [i for i in range(20) if i % 2 ==0]
b

# 代码执行顺序：for —> if —> 表达式



# 筛选L开头的字母
name = ['Lisa', 'Mike', 'Jessica', 'Liming', 'Wangfeng']
name_L = [name[i] for i in range(len(name)) if name[i][0] =='L']
name_L
```

```python
# 循环的嵌套

matric = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matric_1 = [a for b in matric for a in b]
matric_1

# 相当于
h = []
for b in matric:
    for c in b:
        h.append(c)
h



# 嵌套

b = [[m,n] for m in range(10) if m % 2 ==0 for n in range(10) if n % 3 ==0]
b
```

## 函数

```python
# 创建函数

def myfunc(name, times=1):   # times=1 对形参设置默认值，也可以不设置
    if times == 0:
        return 'Error'  # 一旦执行return 则立即退出函数，不会继续执行后面的程序
    else:
        for i in range(times):
            print(name)

# 函数的调用
myfunc('python', 3)  # 利用位置参数，参数的位置必须与定义一致
print()
myfunc(times = 4, name = 'python')  # 利用关键字参数调用函数，就不用在乎参数位置啦
```

### 收集参数类似print可以传入不限定个数的参数

```python
def myfunc(*a):   # *a是指将a定义为一个元组 ，**a 则是将a定义为一个字典
    print('一共传入了{}个参数'.format(len(a)))
    print('第二个参数是{}'.format(a[1]))
    
myfunc('aaa', 'bbbb ', 'cccc')
    
```

```python
# 打包参数

def myfunc(a, *b, **c):    # 按位置参数赋值，注意字典的赋值方法：x=6, y=7, z=8
    print(a, b, c)
    
myfunc(1, 2, 3, 4, 5, x=6, y=7, z=8)  # 赋值结果：a=1, b = (2, 3, 4, 5) , c = {'x': 6, 'y': 7, 'z': 8}
```

```python
# 解包参数

a = (1, 2, 3, 4)
b = {'a':5, 'b':6, 'c':7, 'd':8}

def myfunc(a, b, c, d):
    print(a, b, c, d)
    
myfunc(*a)
myfunc(**b)
    
    
    
    
# global:在函数内部定义全局变量
# nonlocal:内层函数可以改变外层函数的变量值
# 函数只有在定义和调用的时候才会需要小括号
```

### 闭包

```python
def aa(x):
    def bb(y):
        return y ** x
    return bb

cc = aa(3)
dd = aa(2)
print(cc(2), dd(2))
```

```python
# 有记忆函数

def outer():
    a, b=0,0
    def inner(x, y):
        nonlocal a, b
        a+=x
        b+=y
        print(a, b)
    return inner

move = outer()

move(1,2)
move(1,2)  # move两次，因为nonlocal，第一次的结果会影响第二次
```
